```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
position = 0
walk = [position]
nsteps = 1000
for _ in range(nsteps):
    step = 1 if random.randint(0,1) else -1
    position += step
    walk.append(position)
```

```python
plt.plot(walk[:100])
```

```python
# First crossing time: step at which the random walk reaches a particular value
# Also abandoning the pure python implementation in favor of numpy

nsteps = 1000
rng = np.random.default_rng(seed=12345)
draws = rng.integers(0, 2, size=nsteps)
steps = np.where(draws==0,1,-1)
walk = steps.cumsum()

(np.abs(walk) >= 10).argmax() # Returns the first index of max value in the boolean array
```

```python
plt.plot(walk)
```

```python
"""
Can also simulate n random walks at once through modifying the code
"""

nwalks = 500
rng = np.random.default_rng(seed=12345)
draws = rng.integers(0, 2, size=(nwalks, nsteps))
steps = np.where(draws > 0,1,-1)
walks = steps.cumsum(axis=1)
```

```python
plt.figure(figsize=(10, 6))
for i in range(nwalks):
    plt.plot(walks[i], alpha=0.5, linewidth=0.8)

plt.xlabel('Step')
plt.ylabel('Position')
plt.title(f'{nwalks} Random Walks')
plt.grid(True, alpha=0.3)
plt.show()
```
